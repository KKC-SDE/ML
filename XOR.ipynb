```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Original XOR dataset in 2D
x = np.array([[1,1], [-1, -1], [1, -1], [-1, 1]])
y = np.array([1, 1, -1, -1])

# Plotting the 2D points
plt.figure(figsize=(8, 4))
for i, label in enumerate(y):
    plt.plot(x[i, 0], x[i, 1], 'ro' if label == 1 else 'bo')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('2D XOR Dataset')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Original XOR dataset in 2D
x = np.array([[1,1], [-1, -1], [1, -1], [-1, 1]])
y = np.array([1, 1, -1, -1])

# Applying the feature transformation to 3D: x3 = x1 * x2
x_transformed = np.hstack((x, (x[:, 0] * x[:, 1]).reshape(-1, 1)))

# Create a Plotly figure
fig = go.Figure()

# Add the points to the plot
colors = ['red' if label == 1 else 'blue' for label in y]
fig.add_trace(go.Scatter3d(
    x=x_transformed[:, 0],
    y=x_transformed[:, 1],
    z=x_transformed[:, 2],
    mode='markers',
    marker=dict(size=5, color=colors),
))

# Define the separating plane
xx, yy = np.meshgrid(np.linspace(-2, 2, 50), np.linspace(-2, 2, 50))
zz = 0 * xx
fig.add_trace(go.Surface(x=xx, y=yy, z=zz, opacity=0.5, colorscale='Greens'))

# Set the layout for a better view
fig.update_layout(title='Transformed XOR points in 3D with separator',
                  scene=dict(xaxis_title='X1',
                             yaxis_title='X2',
                             zaxis_title='X1 * X2'),
                  margin=dict(r=0, b=0, l=0, t=30))
fig.show()
```
